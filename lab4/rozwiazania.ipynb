# Funkcje pomocnicze do zadaÅ„ 4 i 5

```python
#
# 2d Discrete Cosinus Transform
#
def dct2(array):
    return dct(dct(array, axis=0, norm='ortho'), axis=1, norm='ortho')


def idct2(array):
    return idct(idct(array, axis=0, norm='ortho'), axis=1, norm='ortho')


#
# Calculate quantisation matrices
#
# Based on: https://www.hdm-stuttgart.de/~maucher/Python/MMCodecs/html/jpegUpToQuant.html
#           #step-3-and-4-discrete-cosinus-transform-and-quantisation
#
_QY = numpy.array([[16, 11, 10, 16, 24, 40, 51, 61],
                   [12, 12, 14, 19, 26, 48, 60, 55],
                   [14, 13, 16, 24, 40, 57, 69, 56],
                   [14, 17, 22, 29, 51, 87, 80, 62],
                   [18, 22, 37, 56, 68, 109, 103, 77],
                   [24, 35, 55, 64, 81, 104, 113, 92],
                   [49, 64, 78, 87, 103, 121, 120, 101],
                   [72, 92, 95, 98, 112, 100, 103, 99]])

_QC = numpy.array([[17, 18, 24, 47, 99, 99, 99, 99],
                   [18, 21, 26, 66, 99, 99, 99, 99],
                   [24, 26, 56, 99, 99, 99, 99, 99],
                   [47, 66, 99, 99, 99, 99, 99, 99],
                   [99, 99, 99, 99, 99, 99, 99, 99],
                   [99, 99, 99, 99, 99, 99, 99, 99],
                   [99, 99, 99, 99, 99, 99, 99, 99],
                   [99, 99, 99, 99, 99, 99, 99, 99]])


def _scale(QF):
    if QF < 50 and QF >= 1:
        scale = numpy.floor(5000 / QF)
    elif QF < 100:
        scale = 200 - 2 * QF
    else:
        raise ValueError('Quality Factor must be in the range [1..99]')

    scale = scale / 100.0
    return scale


def QY(QF=85):
    return _QY * _scale(QF)


def QC(QF=85):
    return _QC * _scale(QF)
```

```python
#
# 0. Image data
#
# TODO: implement (zad. 4)
image = numpy.array(numpy.zeros((128, 128, 3)), dtype=numpy.uint8) 

image_size = image.shape[0] * image.shape[1]
image_portion = (image.shape[0] * image.shape[1]) / 7
count = 0

r = 0
g = 0
b = 0

change_value = 255 / image_portion

for x in range(len(image[0])):
    for y in range(len(image)):
        if count <= image_portion:
            b += change_value            
        elif count > image_portion and count <= 2 * image_portion:
            g += change_value
        elif count > 2 * image_portion and count <= 3 * image_portion:
            b -= change_value
        elif count > 3 * image_portion and count <= 4 * image_portion:
            r += change_value
        elif count > 4 * image_portion and count <= 5 * image_portion:
            g -= change_value
        elif count > 5 * image_portion and count <= 6 * image_portion:
            b += change_value
        elif count > 6 * image_portion and count <= 7 * image_portion:
            g += change_value
                            
        if b > 255:
            b = 255
        if g > 255:
            g = 255
        if r > 255:
            r = 255
        if b < 0:
            b = 0
        if g < 0:
            g = 0
        if r < 0:
            r = 0
        
        image[y][x] = (r, g, b)       
        count += 1

#
# 1. Convert RGB to YCbCr
#
# TODO: implement (zad. 4)
YCrCb_image = cv2.cvtColor(image, cv2.COLOR_RGB2YCrCb)

#
# 2. Downsampling on Cb and Cr channels
#
# TODO: implement (zad. 4)
Cb_image = YCrCb_image[:, :, 1]
Cr_image = YCrCb_image[:, :, 2]

scale_percent = 50 
width = int(Cb_image.shape[1] * scale_percent / 100)
height = int(Cb_image.shape[0] * scale_percent / 100)
dim = (width, height)

Cb_image = cv2.resize(Cb_image, dim, interpolation = cv2.INTER_AREA)
Cr_image = cv2.resize(Cr_image, dim, interpolation = cv2.INTER_AREA)

#
# 3. Produce 8x8 blocks
#
# TODO: implement (zad. 4)
Y_channel = YCrCb_image[:, :, 0]
Cr_channel = Cr_image
Cb_channel = Cb_image

Y_blocks = []
Cr_blocks = []
Cb_blocks = []

for i in range(0, Y_channel.shape[0], 8):
    for j in range(0, Y_channel.shape[1], 8):
        Y_blocks.append(Y_channel[i:i + 8, j:j + 8])
        
for i in range(0, Cr_channel.shape[0], 8):
    for j in range(0, Cr_channel.shape[1], 8):
        Cr_blocks.append(Cr_channel[i:i + 8, j:j + 8])
        
for i in range(0, Cb_channel.shape[0], 8):
    for j in range(0, Cb_channel.shape[1], 8):
        Cb_blocks.append(Cb_channel[i:i + 8, j:j + 8])

#
# 4. Calculate DCT on each block
#
# TODO: implement (zad. 5)
for block in Y_blocks:
    block = dct2(block)
    
for block in Cr_blocks:
    block = dct2(block)
    
for block in Cb_blocks:
    block = dct2(block)

#
# 5. Divide each block by quantisation matrix
#
# TODO: implement (zad. 5)
my_qy = QY(50)

for block in Y_blocks:
    block = block / my_qy
    
for block in Cr_blocks:
    block = block / my_qy
    
for block in Cb_blocks:
    block = block / my_qy

#
# 6. Round values in each block to integers
#
# TODO: implement (zad. 5)
for block in Y_blocks:
    for i in range(len(block)):
        for j in range(len(block[i])):
            block[i][j] = int(round(block[i][j]))
            
for block in Cr_blocks:
    for i in range(len(block)):
        for j in range(len(block[i])):
            block[i][j] = int(round(block[i][j]))
            
for block in Cb_blocks:
    for i in range(len(block)):
        for j in range(len(block[i])):
            block[i][j] = int(round(block[i][j]))

#
# 7. Zig Zag
#
# TODO: implement (zad. 4)
w = len(Y_blocks)
h = 64
Y_after_zig_zag = [[0 for x in range(h)] for y in range(w)]

count = 0
for block in Y_blocks:    
    
    solution=[[] for i in range(8 + 8 - 1)]
    
    for i in range(8):
        for j in range(8):
            sum=i+j
            if(sum%2 ==0):  
                solution[sum].insert(0, block[i][j])
            else:           
                solution[sum].append(block[i][j])
                
    count2 = 0
    
    for i in solution:
        for j in i:
            Y_after_zig_zag[count][count2] = j
            count2 += 1
            
    count += 1

#################################################################
w = len(Cr_blocks)
h = 64
Cr_after_zig_zag = [[0 for x in range(h)] for y in range(w)]
    
count = 0
for block in Cr_blocks:    
    
    solution=[[] for i in range(8 + 8 - 1)]
    
    for i in range(8):
        for j in range(8):
            sum=i+j
            if(sum%2 ==0):  
                solution[sum].insert(0, block[i][j])
            else:           
                solution[sum].append(block[i][j])
                
    count2 = 0
    
    for i in solution:
        for j in i:
            Cr_after_zig_zag[count][count2] = j
            count2 += 1
            
    count += 1

#################################################################
w = len(Cb_blocks)
h = 64
Cb_after_zig_zag = [[0 for x in range(h)] for y in range(w)]
    
count = 0
for block in Cb_blocks:    
    
    solution=[[] for i in range(8 + 8 - 1)]
    
    for i in range(8):
        for j in range(8):
            sum=i+j
            if(sum%2 ==0):  
                solution[sum].insert(0, block[i][j])
            else:           
                solution[sum].append(block[i][j])
                
    count2 = 0
    
    for i in solution:
        for j in i:
            Cb_after_zig_zag[count][count2] = j
            count2 += 1
            
    count += 1


#
# 8. Flatten, concatenate, compress and calculate the size -- how many bytes?
#
# TODO: implement (zad. 4)
combined_channels = []

flat_Y = [item for sublist in Y_after_zig_zag for item in sublist]
flat_Cr = [item for sublist in Cr_after_zig_zag for item in sublist]
flat_Cb = [item for sublist in Cb_after_zig_zag for item in sublist]

for item in flat_Y:
    combined_channels.append(item)
for item in flat_Cr:
    combined_channels.append(item)
for item in flat_Cb:
    combined_channels.append(item)
    
compressed = zlib.compress(bytes(combined_channels))
#print(len(compressed))

#
# 7'. Undo Zig Zag
#
# We can skip it in this exercise! We did Zig Zag only for analysis in step 8.
# You can continue with result from step 6. instead of implementing undo here.
#

#
# 6'. Nothing to do here   Â¯\_(ãƒ„)_/Â¯
#
# No conversion is really needed here, just proceed to the next step.
#

#
# 5'. Reverse division by quantisation matrix -- multiply
#
# TODO: implement (zad. 5)
my_qc = QC(50)

for block in Y_blocks:
    block = block * my_qc
    
for block in Cr_blocks:
    block = block * my_qc
    
for block in Cb_blocks:
    block = block * my_qc

#
# 4'. Reverse DCT
#
# TODO: implement (zad. 5)
for block in Y_blocks:
    block = idct2(block)
    
for block in Cr_blocks:
    block = idct2(block)
    
for block in Cb_blocks:
    block = idct2(block)


#
# 3'. Combine 8x8 blocks to original image
#
# TODO: implement (zad. 4)
step = int(Y_channel.shape[0] / 8)
rows = []
for i in range(0, len(Y_blocks), step):
    rows.append(numpy.concatenate(Y_blocks[i:i + step], axis=1))
Y_channel = numpy.concatenate(rows, axis=0)

step = int(Cr_channel.shape[0] / 8)
rows = []
for i in range(0, len(Cr_blocks), step):
    rows.append(numpy.concatenate(Cr_blocks[i:i + step], axis=1))
Cr_channel = numpy.concatenate(rows, axis=0)


step = int(Cb_channel.shape[0] / 8)
rows = []
for i in range(0, len(Cb_blocks), step):
    rows.append(numpy.concatenate(Cb_blocks[i:i + step], axis=1))
Cb_channel = numpy.concatenate(rows, axis=0)

#
# 2'. Upsampling on Cb and Cr channels
#
# TODO: implement (zad. 4)
scale_percent = 200 
width = int(Cr_channel.shape[1] * scale_percent / 100)
height = int(Cb_channel.shape[0] * scale_percent / 100)
dim = (width, height)  

Cb_image = cv2.resize(Cb_image, dim, interpolation = cv2.INTER_AREA)
Cr_image = cv2.resize(Cr_image, dim, interpolation = cv2.INTER_AREA)

new_YCrCb = image.copy()

new_YCrCb[:, :, 0] = Y_channel
new_YCrCb[:, :, 1] = Cb_image
new_YCrCb[:, :, 2] = Cr_image

# 1'. Convert YCbCr to RGB
#
# TODO: implement (zad. 4)
new_RGB = cv2.cvtColor(new_YCrCb, cv2.COLOR_YCrCb2RGB)

#
#
# 0'. Save the decoded image -- as PPM or PNG
#
# TODO: implement (zad. 4)
ppm_ascii_header = f'P3 {128} {128} {255}\n'

with open('lab4-po-kompresji-i-dekompresji-jpeg.ppm', 'w') as fh:
    fh.write(ppm_ascii_header)
    new_RGB.tofile(fh, sep=' ')
    fh.write('\n')
```
